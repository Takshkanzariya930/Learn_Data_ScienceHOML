```python
import matplotlib.pyplot as plt

# Data
cm = [1, 2, 3, 4, 5, 6]
gm = [0.0254, 0.0192, 0.0105, 0.0154, 0.0154, 0.0159]
thita = [0.832, 2.234, 3.806, 5.572, 6.952, 8.051]

# Create the plot
plt.figure(figsize=(8, 5), dpi=120)

# Plot gm
plt.plot(cm, gm, marker='o', linestyle='-', color='tab:blue',
         label=r'$g_m$', linewidth=2, markersize=8)

# Plot thita
plt.plot(cm, thita, marker='s', linestyle='--', color='tab:orange',
         label=r'$\theta$', linewidth=2, markersize=8)

# Beautify
plt.title("Variation of $g_m$ and $\\theta$ with cm", fontsize=14, fontweight="bold")
plt.xlabel("cm", fontsize=12)
plt.ylabel("Values", fontsize=12)

plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=11)
plt.xticks(cm)  # Ensure x-axis ticks match cm values
plt.tight_layout()

# Show the plot
plt.show()
```
